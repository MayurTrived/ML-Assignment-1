```python
# 8  Write a Python script to:
# Generate 1000 samples from a normal distribution.
# Introduce 50 random missing values.
# Fill missing values with the column mean.
# Plot a histogram before and after imputation.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(50, 10, 1000)

missing_indices = np.random.choice(1000, 50, replace=False)
data[missing_indices] = np.nan

plt.hist(data, bins=30)
plt.title('Before Imputation')
plt.show()

imputed_data = pd.Series(data).fillna(np.nanmean(data))

plt.hist(imputed_data, bins=30)
plt.title('After Imputation')
plt.show()
```

```python
# 9  Implement Min-Max scaling on the following list of numbers [2, 5, 10, 15,20] using sklearn.preprocessing.MinMaxScaler. Print the scaled array.

from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler()
data = np.array([[2], [5], [10], [15], [20]])
scaled = scaler.fit_transform(data)
print(scaled)
```
